Let's start off by making some pretty graphics with the [Matplotlib](http://matplotlib.org/) library. If you have ever used R, MATLAB or Mathematica, using this might feel familiar. This package tries to make data plotting as painless as possible, and comes with no strings attached. If you want more control over your graphics, check out the list of [Plotting libraries](https://wiki.python.org/moin/NumericAndScientific/Plotting) on python.org. We are going to now add Matplotlib and its external links to our environment, then draw a chart from some open data.

````python
%matplotlib inline
````

Note: you can also use the [GUI backends](http://matplotlib.org/faq/usage_faq.html#what-is-a-backend) here, e.g. `%matplotlib qt`, if you prefer and your system supports them. Also note, that to some people this reliance on external dependencies is exactly the reason why some people choose to *not* use Matplotlib, as it makes your code less portable. Pick your battles.

Now we import into our namespace:

````python
import matplotlib.pyplot as plt
import numpy as np
````

Here is a six second beat pattern of the Mountain Bluebird (Sialia currucoides, which [resembles](http://www.nytimes.com/2014/08/10/magazine/who-made-that-twitter-bird.html?_r=0) the Twitter logo), extracted using [Audacity](http://audacity.sourceforge.net/) from a recording obtained on [xeno-canto](http://www.xeno-canto.org/species/Sialia-currucoides) (Creative Commons license BY-NC-SA). I'm sure it is possible to do this in Python as well, but it's hard to replace a good tool once it's already in your belt.

The resulting rhythm is a series of numbers that can be just copy-pasted into a Python array. Transforming into a NumPy array gives our bird-song superpowers. Or at least makes it a lot more efficient to process later on:

````python
birdsong_data = [0.080000, 0.444000, 0.448000, 0.449000, 0.667000, 0.670000, 0.932000, 0.938000, 1.384000, 2.338000, 2.428000, 2.434000, 3.979000, 4.043000, 4.094000, 4.100000, 4.174000, 4.301000, 4.303000, 4.311000, 4.421000, 4.631000, 4.833000, 4.961000, 4.965000, 5.094000, 5.097000, 5.178000, 5.183000, 5.473000, 5.478000, 5.766000, 5.767000, 5.772000, 5.841000, 5.843000]
birdsong = np.array(birdsong_data)
````

We can plot this sequence in time with the handy [scatter](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter) graph:

````python
plt.scatter(birdsong, [1]*len(birdsong), s=50, alpha=0.5)
plt.title('A scatter chirp-o-gram')
````

Or in linear space using the ubiquitous [plot](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) function.

````python
x = np.linspace(0, 6, len(birdsong))
plt.plot(x, birdsong)
plt.title('A simple chirp-o-gram');
````

And that's like the little snow-heap at the top of the iceberg of the kinds of charts you can do with Python. Then again, if you're planning to publish your data on the Web a better option might be to use something like [D3.js](http://d3js.org/) or [Raphael.js](http://raphaeljs.com/) for crispness and interactivity. ## Velo flows

As you may know, [Stadt Zürich has a portal](http://data.stadt-zuerich.ch) that puts together data from all kinds of departments and makes it more accessible to the general public. Let's use this to learn something about this lovely city. Do you like riding your bike through town - or maybe you would like to know just where all those cyclists are rushing off to in the morning? 

You can get readings of traffic measured as bikes pass through special lanes in the neat collection [Daten der permanenten Velozählstellen](http://data.stadt-zuerich.ch/portal/de/index/ogd/daten/velozaehlstellen.html). This data has already been explored by my esteemed colleague Dr. Ralph Straumann using R in a series of blog posts: [Teil 1](http://geo.ebp.ch/2014/02/18/blick-in-die-werkzeugkiste-offene-daten-in-r-teil1/), [Teil 2](http://geo.ebp.ch/2014/04/04/blick-in-die-werkzeugkiste-offene-daten-in-r-teil-2/), [Teil 3](http://geo.ebp.ch/2014/04/25/blick-in-die-werkzeugkiste-offene-daten-in-r-teil-3/). We're not going to go into as much detail here, but let's see how Python with Pandas compares for processing these 30 MB of open data in CSV format.

Start by importing Pandas as we did with Matplotlib:

````python
import pandas as pd
````

We can now read in the CSV file we downloaded - you could just as well grab it directly from the Web:

````python
bike_data = pd.read_csv('/opt/localdev/SciPy/data/TAZ_velozaehlung_stundenwerte.csv')
````

Now let's look at a few rows to make sure we loaded what we expected:

````python
# Equivalent to bike_data[-6:-1]
# For the top 5 use bike_data.head(5)
bike_data.tail(5)
````

That was easy! Remember loading CSV by hand? What Pandas just did for us was not only automatically detect the delimiter and read in the entire file, but create an indexed, columnar data structure called a [DataFrame](http://pandas.pydata.org/pandas-docs/stable/dsintro.html#dataframe), which can be accessed by column name and will serve us well in the later analysis. You can tweak a lot of things in this command, such as specifying an alternate primary key or helping the date parser - see [read_csv()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.io.parsers.read_csv.html).

There are some handy functions to get a feeling for the data before digging in:

````python
bike_data.describe()
````

At a glance, we can see that we have a nicely distributed data set, with no apparent gaps - except for anything explicitly labelled as "Ausfall"; using nulls instead of 0's would be better, something Ralph already noted in his blog.

We can also quickly check other basics, like a possible skew in the dataset, with Pandas statistics functions:

````python
bike_data['Stunde'].median()
````

````python
bike_data['Zählstelle (Code)'].unique()
````

Pandas interacts well with Matplotlib, as we shall see next.

## I see data, people!

````python
pd.options.display.mpl_style = 'default' # the 'look nicer' option
plt.scatter(bike_data['Stunde'], bike_data['Gezählte Velofahrten'])
````

Scatterplots are a brute force approach to plotting data. We had a brief conversation about this during the meet-up concerning the relative merits and performance optimization in Python and R. If you care for this kind of thing, consider this [article](http://tullo.ch/articles/speeding-up-isotonic-regression/), and try [Julia](http://julialang.org/). For a faster scatter also see [ggplot](https://github.com/yhat/ggplot). For a prettier scatter go with [seaborn](https://github.com/mwaskom/seaborn).

Here is an actual histogram of actual counts using the `plot()` function through the `DataFrame` itself:

````python
bike_data.groupby(['Stunde'])['Gezählte Velofahrten'].sum().plot(kind='bar')
````

And here we can see the hourly histogram for each recorded station using the `hist()` function:

````python
fig = bike_data['Gezählte Velofahrten'].hist(by=bike_data['Zählstelle (Code)'], grid=False, figsize=(26,26))
````

Hope this has given you a good taster for what Pandas, together with Numpy and Matplotlib, can do for your data. It's an outstanding alternative to expensive proprietary tools, and while perhaps less "intimate" than manipulating numbers in arrays, its a foundation in statistical science that lets us work with data at an undoubtedly faster pace.
